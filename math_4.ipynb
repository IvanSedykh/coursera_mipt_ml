```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

## возьмем экспоненциальное распределние
его плотность: 
$
f(x) = \lambda e^{-\lambda x}
$

```python
lambd = 0.3
beta = 1/lambd
X = np.random.exponential(scale=beta,size=1000) # получим семплы из распределния

# xx, yy = density(lambd) # получим плотность
xx = np.linspace(0, 25, 1000)
yy = stats.expon.pdf(xx,scale=beta)

# нарисуем
plt.hist(X, bins=50, normed=True, label='samples')
plt.plot(xx, yy, label='density')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.show()
```

мат ожидание и дисперися экспоненциального распределения:
$$ E[X] = \frac{1}{\lambda}$$
$$D[X] = \frac{1}{\lambda^2} $$
тогда согласно ЦПТ мат ожидание и дисперсия средних:
$$E[\overline{X_i}] = E[X]$$
$$D[\overline{X_i}] = \frac{D[X]}{n}$$

```python
ns = [3,5,10,25,50,100,300,1000,5000,10000]
for n in ns:
    print('n = {}'.format(n))
    n = 40
    X = np.random.exponential(scale=beta, size=(1000,n))
    means = X.mean(axis=1)

    mu = 1/lambd # матожидание
    sigma = 1/lambd**2 # дисперсия
    xx = np.linspace(0,8,1000)
    yy = stats.norm.pdf(xx, mu, np.sqrt(sigma/n))

    plt.hist(means, bins=20, normed=True, label='sample means(normed)')
    plt.plot(xx,yy, label='normal distr density')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.legend()
    plt.show()
```

### Как видно из графиков, с ростом n качество аппроксимации растет(хотя даже при малых n качество уже неплохое), гистограмма все больше похожа на плотность нормального распределения. Тем самым подтверждается ЦПТ.
