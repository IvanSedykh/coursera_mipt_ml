```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
xx = np.linspace(1,30,60)
plt.plot(xx,f(xx))
plt.show()
```

```python
x_0 = np.array([2])
result = optimize.minimize(f, x_0, method='BFGS',bounds=[(1,30)])
result
```

```python
x_0 = np.array([30])
result = optimize.minimize(f, x_0, method='BFGS',bounds=[(1,30)])
result
```

```python
x_0 = np.array([30])
result = optimize.differential_evolution(f, bounds=[(1,30)])
result
```

```python
def g(x):
    return f(x).astype('int')

xx = np.linspace(1,30,1000)
plt.plot(xx,g(xx))
plt.show()
```

```python
x_0 = np.array([30])
result = optimize.minimize(g, x_0, method='BFGS',bounds=[(1,30)])
result
```

```python
x_0 = np.array([30])
result = optimize.differential_evolution(g, bounds=[(1,30)])
result
```
